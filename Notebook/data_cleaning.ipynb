```python
# Columns with missing values are . 
missing_cols = []
for i in data_import : 
    if data_import[i].isnull().sum() > 0 : 
        missing_cols.append(i)
    else : 
        continue

```

```python
missing_cols
```

```python
from sklearn.impute import SimpleImputer 

imputer = SimpleImputer()


data_import[missing_cols] = imputer.fit_transform(data_import[missing_cols])
```

```python
data_import[missing_cols].isnull().sum()
```

```python
def get_max_min_dict(df , num_col) : 
    result_dict = {}

    for col in num_col : 
        max_val = df[col].max()
        min_val = df[col].min()

        result_dict[col] = (max_val , min_val)

    return result_dict

max_min_dict = get_max_min_dict(data_import , num_col)
max_min_dict
```

```python
# Let's convert Status col first , Developing will be 0 and Developed will be 1. 

data_import['Status'] = data_import['Status'].map({'Developing' : 0 , 'Developed' : 1})
```

```python
data_import['Status'].value_counts()
```

```python
# for Country , let's use LabelEncoder. 

from sklearn.preprocessing import LabelEncoder

le_encoder = LabelEncoder()

data_import['Country'] = le_encoder.fit_transform(data_import['Country'])


with open('../artifacts/country_encoder.pkl' , 'wb') as le_encoding : 
    pickle.dump(le_encoder , le_encoding)
```
